# ML Logistic Regression
- This is a binary case, the output is either Grass or Fire type.
- Adagrad is used as the optimizer.
- I used Sklearn in the end to compare the coefficient and accuracy, finding the coefficients I got is close to the one generated by sklearn model.

```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
data = pd.read_csv('/Users/yuwenchen/Pokemon-Classifier-V2/Pokemon.csv')
data = pd.DataFrame(data)
data.head()
```

```python
df = data[['Type 1', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
```

```python
# extract all grass and fire type pokemon 
grassType = df.loc[df['Type 1'] == 'Grass']
fireType = df.loc[df['Type 1'] == 'Fire']

# let C1(grass) = 1, C2(fire) = 0
grassType['Type'] = 1
fireType['Type'] = 0

# 80% of grass and fire type pokemon as training set
grass_train = grassType.sample(frac=0.8, replace=False, random_state=1)
fire_train = fireType.sample(frac=0.8, replace=False, random_state=1)
all_train = pd.concat([grass_train, fire_train], ignore_index=True, sort=False)

# the rest of the 20% as testing set
grass_test = grassType[~grassType.isin(grass_train)].dropna()
fire_test = fireType[~fireType.isin(fire_train)].dropna()
all_test = pd.concat([grass_test, fire_test], ignore_index=True, sort=False)
```

```python
# shuffle the order
all_train = shuffle(all_train)
```

```python
x_data = all_train[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].values.astype(float).tolist()
y_data = all_train['Type'].astype(float).tolist()
```

```python
all_train.head()
```

# Training Phase
- Model (funtion set) is same as linear regression but adding a sigmoid function so that the output will be limited from 0 to 1.
    - $i$ represents the number of attributes or feature (HP, Attacks, Defense, etc).
    - Therefore, w is a vector $[w_1, w_2, w_3, w_4, w_5, w_6]$.
    - Input $x$ is just one row of data here.
<br><br>
    $$
    f_{w,b}(x) = \sigma(\sum_{i}\ w_i x_i + b)
    $$
<br><br>
- Loss function is the cross entropy between two Bernoulli distribution.
    - $\hat{y}^n$: 1 for class 1 (grass), 0 for class 2 (fire).
    - $n$ represents the number of each data or row
    - $C$ represents Cross Entropy
<br><br>
    $$
    L(f_{w,b}(x)) = \sum_{n}\ C(f_{w,b}(x^n), \hat{y}^n )
    $$
<br><br>
    - Cross Entropy would be like:<br><br>
    $$
    C(f_{w,b}(x^n), \hat{y}^n) = -[y^n \ln  f_{w,b}(x^n) + (1-y^n) \ln (1-f_{w,b}(x^n))]
    $$
<br><br>
- Use gradient descent to find the  best w and b ($w^*, b^*$).
    - After we do the differential, the way we updating the w and b is same as linear regression.
<br><br>
    $$
    w_i = w_i - \eta \sum_{n} -(\hat{y}^n-f_{w,b}(x^n))x^n_i
    $$
    $$
    b = b - \eta \sum_{n} -(\hat{y}^n-f_{w,b}(x^n)) \times 1
    $$
<br><br>
    - Notice that for we have to updats every $w_i$. In this case 6 times.

```python
# fSet is the function set
# x_row is one row with 6 attributes(HP, Attack, Defense, etc.)
def fSet(x_row: list, w: list, b):
    result = 0.0
    for r in range(len(x_row)):
        result = result + (w[r]*x_row[r])
    result = result + b
    return sigmoid(result)

def sigmoid(x):
    sig = 1 / (1 + np.exp(-x))     # Define sigmoid function
    sig = np.minimum(sig, 0.9999)  # Set upper bound
    sig = np.maximum(sig, 0.0001)  # Set lower bound
    return sig
```

```python
# w = [w1, w2, w3, w4, w5, w6]
# w1 for HP, w2 for Attack, w3 for Defense, w4 for Sp. Atk, w5 for Sp. Def, w6 for Speed
att_len = len(x_data[0])

b = 1.0 # initial b
lr = 0.001
w = [0]*6
iteration = 10000

# store vlaues for plotting
b_history = []
w_history = []
loss_history = []

# learning rate
lr_b = 0.001
lr_w = [0.001]*att_len

for i in range(iteration):
    b_grad = 0.0
    w_grad = [0.0]*att_len
    loss = 0.0
    
    # caculating the loss of each iteration
    # for current 'w' and 'b' what is its loss
    for n in range(len(x_data)):
        
        # Cross Entropy to compute the loss 
        loss = loss + (-1)*((y_data[n]*np.log(fSet(x_data[n], w, b))) + ((1 - y_data[n])*np.log(1-fSet(x_data[n], w, b))))
        
        # compute ∂L/∂w1, ∂L/∂w2, ∂L/∂w3, ∂L/∂w4, ∂L/∂w5, ∂L/∂w6
        for m in range(att_len):
            w_grad[m] = w_grad[m] - (y_data[n] - fSet(x_data[n], w, b))*x_data[n][m]
        
        # compute ∂L/∂b 
        b_grad = b_grad - (y_data[n] - fSet(x_data[n], w, b))*1.0
    
    # updats b and all the w
    # use Adagrad for leanring rate
    lr_b = lr_b + b_grad ** 2
    for k in range(att_len):
        lr_w[k] = lr_w[k] + w_grad[k] ** 2
    b = b - lr/np.sqrt(lr_b) * b_grad
    for j in range(att_len):
        w[j] = w[j] - lr/np.sqrt(lr_w[j]) * w_grad[j]
    
    if i % 1000 == 0:
        print("Step: {}. Loss: {}".format(i, loss))
    
    loss_history.append(loss)
    b_history.append(b)
    w_history.append(w)
```

# Testing Phase

```python
all_test
```

```python
fig = plt.figure()
plt.title("Loss")
plt.plot(range(0,iteration), loss_history[0:iteration])    
plt.show()
```

```python
test = all_test.values.T.tolist()
tlen = all_test.shape[0]
```

```python
TP = TN = FP = FN = 0
for i in range(tlen):
    temp = [int(test[1][i]) ,int(test[2][i]), int(test[3][i]), int(test[4][i]), int(test[5][i]), int(test[6][i])]
    pro = fSet(temp ,w, b)
    
    if pro > 0.5: # the prediction is grass
        if test[0][i] == 'Grass': # if it is actually grass type
            TP += 1
        else: # if it is actually NOT Grass type
            FP += 1
    else: # the prediction is fire
        if test[0][i] == 'Fire': # if it is actually fire type
            TN += 1
        else: # if it is actually NOT fire type
            FN += 1
```

```python
accuracy = (TP+TN) / (TP+TN+FP+FN)
accuracy
```

# Use Sklearn to verify my own model

```python
from sklearn.linear_model import LogisticRegression

x_train = all_train[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
y_train = all_train[['Type']]

model=LogisticRegression(max_iter = 10000)
model.fit(x_train, y_train)
```

```python
x_test = all_test[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
y_test = all_test[['Type']]

model.score(x_test, y_test)
```
